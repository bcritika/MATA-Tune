```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
```

```python
options = Options()
options.add_argument('--headless')  # Run Chrome in headless mode (no browser UI)
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-blink-features=AutomationControlled')  # Bypass bot detection
options.add_argument('--remote-debugging-port=9222')  # Helps in some loading issues

# Initialize the Chrome driver with webdriver-manager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
try:
    driver.get("https://www.investors.com/")
    print("Page loaded successfully")
except Exception as e:
    print("Failed to load page:", e)

# Wait for the page to load
time.sleep(5)  # Adjust as needed to give the page time to load

soup = BeautifulSoup(driver.page_source, 'html.parser')

print(soup)
# Find the "Stocks On The Move" section
moved_stocks_section = soup.find('div', class_='moved-stocks')

# If the section is found, print or process it
if moved_stocks_section:
    print("Stocks On The Move section found")
else:
    print("Could not find the 'Stocks On The Move' section")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Set up the options for headless browsing
options = Options()
options.headless = True

# Set the path to your web driver (ChromeDriver for Chrome)
driver = webdriver.Chrome(options=options)

# Visit the webpage
driver.get("https://www.investors.com/")

# Wait for the page to load
driver.implicitly_wait(10)

# Now you can search for the moved-stocks class
soup = BeautifulSoup(driver.page_source, 'html.parser')

print(soup)
# Find the "Stocks On The Move" section
moved_stocks_section = soup.find('div', class_='moved-stocks')

# If the section is found, print or process it
if moved_stocks_section:
    print("Stocks On The Move section found")
else:
    print("Could not find the 'Stocks On The Move' section")

driver.quit()
```
